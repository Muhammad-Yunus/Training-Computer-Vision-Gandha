# 4.1 Into OpenCV | 25+ Tahun Perkembangan OpenCV
<img src="res/opencv.png" width=150><br>
#### Era Intel (1999–2008)
- **1999** — OpenCV pertama kali dikembangkan oleh **Intel** sebagai library open-source untuk computer vision.
- **2000** — Versi *alpha* dirilis ke publik pada konferensi **IEEE CVPR**.
- **2006** — Rilis **OpenCV 1.0** (berbasis C).  
- **2008** — Rilis **OpenCV 1.1 "pre-release"**.
- **2008** — Pengembangan OpenCV dialihkan ke **Willow Garage** & **Itseez** (developer utama).  

#### Era Willow Garage / Itseez (2008–2012)
- **Oktober 2009** — Rilis **OpenCV 2.0**  
  (major update: API C++ modern, Python bindings, struktur modul baru).
- **2010–2012** — Pengembangan pesat, termasuk fitur stitching, object detection modern, GPU (CUDA), dsb.
- **2012** — Manajemen proyek dipindahkan ke **OpenCV.org Foundation**.  
  Repository resmi **migrasi dari SVN ke GitHub**.  

#### Era OpenCV Foundation & Intel Reacquisition (2012–sekarang)
- **Juni 2015** — Rilis **OpenCV 3.0**  
  (T-API OpenCL acceleration, modular API, performance upgrade).  
- **2016** — **Intel mengakuisisi Itseez**, sehingga OpenCV kembali berada di bawah Intel.  
- **Akhir 2016** — Rilis **DNN module** (deep learning inference) dan **OpenCV.js** (JavaScript/WebAssembly).  
  Mendukung model Caffe, Torch, ONNX, TensorFlow.

#### Integrasi AI & Ekosistem Hardware
- **2018** — Intel merilis **OpenVINO Toolkit**, menjadi salah satu backend utama untuk OpenCV DNN.  
- **Nov 2018** — Rilis **OpenCV 4.0**  
  (C++11, build system modern, optimasi CPU/GPU, Vulkan backend).
- **2020 Juli** — Rilis **OpenCV 4.4**.
- **2020 Juli** — OpenCV bekerja sama dengan Luxonis merilis **OAK (OpenCV AI Kit)**:  
  kamera edge-AI dengan DepthAI (spatial AI + depth stereo).

#### Perkembangan 2021–2025
- **2021** — Rilis seri **OpenCV 4.5.x**, dengan peningkatan DNN (ONNX ops baru, INT8 quantization, MobileNet optimizations).
- **2021** — OpenCV AI Kit **OAK-D Lite** diluncurkan, memperluas ekosistem spatial AI.
- **2022** — Rilis **OpenCV 4.6** dan **4.7**  
  (perbaikan pada G-API, dukungan hardware acceleration lebih baik).
- **2023** — Rilis **OpenCV 4.8**  
  (peningkatan ONNX DNN backend, update ke zlib/libpng/libjpeg-turbo).
- **2024** — Rilis **OpenCV 4.9**  
  (OpenCL modernized, perbaikan DNN, dukungan transformer-based ops pada backend ONNX).
- **2024 akhir** — Stabilitas API 4.x, eksperimen modul **OpenCV 5 alpha** dimulai (private branch).
- **2025** *(perkiraan/realitas community status)*  
  - Pengembangan menuju **OpenCV 5.0**:  
    - API C++17  
    - Kernel acceleration unified untuk CPU/GPU/Vulkan  
    - Update penuh DNN (ops transformer, ONNX 1.16+)  
    - Peningkatan G-API & plugin graph runtimes  
  - Ekosistem OAK/DepthAI semakin terintegrasi ke pipeline OpenCV.
