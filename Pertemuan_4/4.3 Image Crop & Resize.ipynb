## 4.3.1 Image Crop
- We can use *array slicing* to Crop Image Matrix<br><br>
- using this format `image_array[y_min:y_max , x_min:x_max]`<br><br>
- where `y_min`, `y_max`, `x_min` and `x_max` is pixel coordinate where the image cropped.<br><br>
    <img src="res/crop_img.png" style="width: 400px;"></img><br><br>
    <img src="res/crop_image_il.png" style="width: 400px;"></img>

```python
# read image from file 'lena.jpg'
img = cv2.imread('lena.jpg')
```

```python
# check shape of the image in form of (height, width, channels)
img.shape
```

```python
# crop image [y_min:y_max , x_min:x_max]
# crop to size 150x150 pixels from top-left corner

img_crop = img[0:150, 0:150] # y: 0 to 150, x: 0 to 150
```

```python
img_crop.shape
```

```python
# show image original and croped image
cv2.imshow('croped image',img_crop)
cv2.imshow('original image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 4.3.2 Image Resize

- OpenCV function to resize image :<br><br>
    - `cv2.resize(img, (w_new, h_new))` : resize `img` to dimension `w_new` x `h_new`<br><br>
    <img src="res/resize.jpg" style="width: 600px;"></img>

```python
img.shape
```

```python
# read image from file 'lena.jpg'
img = cv2.imread('lena.jpg')


# new dimension specified as tuple (new_width, new_height)
# e.g. resize image to 140x140 pixels and store it to variable img_resize
img_resize = cv2.resize(img, (140, 140))  


# show original and resized image in separate windows
cv2.imshow('Original Image', img)
cv2.imshow('Resized Image', img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- resize using `fx` and `fy` parameter,<br><br>
    - `cv2.resize(img, (0, 0), fx=ratio, fy=ratio)`

```python
# read image from file 'lena.jpg'
img = cv2.imread('lena.jpg')

# get original image dimensions
h, w, c = img.shape

# set resize ratio
# value between 0 and 1 for downscaling
# > 1 for upscaling
ratio = 0.5


# resize image using scaling factors, 
# fx for width, fy for height
# ignore (0,0) parameter since we have fx and fy,
# no need to calculate new dimensions manually
img_resize = cv2.resize(img, (0,0), fx=ratio, fy=ratio)  


# show image original and resized image
cv2.imshow('Original Image', img)
cv2.imshow('Resized Image', img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
