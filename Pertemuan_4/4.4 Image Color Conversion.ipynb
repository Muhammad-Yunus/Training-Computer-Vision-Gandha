## 4.1 OpenCV Implementation

```python
# read image from file 'lena.jpg'
img = cv2.imread('lena.jpg')


# convert BGR to Gray
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Original', img)
cv2.imshow('Gray', img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

- Image to gray conversion often used in preprocessing stage of computer vision tasks
- to reduce computational complexity while retaining essential information.
- Gray images have only one channel compared to three channels in color images (BGR),
- leading to faster processing times and lower memory usage.
- in openCV, many algorithms such as edge detection, thresholding, and feature extraction operate on gray images.
