## 2.2 Python Function

<font color="orange">**Function**</font> is a block of code that can be reused (<font color="orange">**reusable**</font>). <font color="orange">**Module**</font> is a Python file that contains a collection of functions and <font color="orange">**variables**</font>.

<br><br><br><br>
___

#### 2.2.1 Defining a Function

To create a function, use the <font color="orange">**def**</font> keyword.

**Syntax:**
```python
def function_name():
    # code inside function
```

````python
# Example 2.1.1: Defining a Simple Function
def greet():
    print("Hello, Welcome!")
````

````python
# Calling the function
greet()
````

<br><br>
#### 2.2.2 Function with Parameters

<font color="orange">**Parameter**</font> is a variable passed to a function when it is called. Use <font color="orange">**argument**</font> (value) to fill the parameter.

````python
# Example 2.1.2: Function with Parameters
def greet_name(name):
    print("Hello", name + "!")
````

````python
# Calling the function with arguments
greet_name("Budi")
````

````python
# Calling the function with arguments
greet_name("Ani")
````

<br><br>
#### 2.2.3 Function with Multiple Parameters

A function can have more than one <font color="orange">**parameter**</font>. Separate parameters with a comma.

````python
# Example 2.1.3: Function with Multiple Parameters
def add(a, b):
    result = a + b
    print(a, "+", b, "=", result)
````

````python
# Calling the function

add(10, 20)
````

````python
# Calling the function

add(5, 3)
````

<br><br>
#### 2.2.4 Return Statement

The <font color="orange">**return**</font> keyword is used to return a value from a function. The returned value can be stored in a variable.

````python
# Example 2.1.4: Function with Return
def calculate_square_area(side):
    area = side * side
    return area
````

````python
# Calling the function and storing the result

result = calculate_square_area(5)

print(result)
````

````python
# Calling the function and storing the result

result = calculate_square_area(10)

print(result)
````

<br><br>
#### 2.2.5 Function with Conditional

A function can also use <font color="orange">**conditional statements**</font> (if, elif, else) inside it.

````python
# Example 2.1.5: Function with Conditional
def check_number(number):
    if number > 0:
        return "Positive number"
    elif number < 0:
        return "Negative number"
    else:
        return "Zero"
````

````python
# Calling the function and printing the result

result = check_number(5)

print(result)
````

````python
# Calling the function and printing the result

result = check_number(-3)

print(result)
````

````python
# Calling the function and printing the result

result = check_number(0)

print(result)
````

<br><br>
#### 2.2.6 Function with Loop

A function can also use <font color="orange">**loops**</font> (while or for) inside it.

````python
# Example 2.1.6: Function with Loop
def calculate_total(n):
    total = 0
    for i in range(1, n + 1):
        total = total + i
    return total
````

````python
# Calling the function and storing the result

result = calculate_total(5)

print(result)
````

````python
# Calling the function and storing the result

result = calculate_total(9)

print(result)
````

<br><br>
#### 2.2.7 Calling Function Inside Another Function

Functions can call other functions. This helps break down complex problems into smaller, manageable pieces.

````python
# Example 2.2.7: Function calling another function
def calculate_square(number):
    result = number * number
    return result

def calculate_sum_of_squares(a, b):
    square_a = calculate_square(a)
    square_b = calculate_square(b)
    total = square_a + square_b
    return total
````

````python
# Calling the function that uses another function inside

result = calculate_sum_of_squares(3, 4)
print("3² + 4² =", result)
````

````python
# Another example call

result = calculate_sum_of_squares(5, 12)
print("5² + 12² =", result)
````
