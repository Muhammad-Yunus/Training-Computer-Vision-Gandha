## 3. Module

<font color="orange">**Module**</font> is a Python file (.py) that contains functions and variables that can be used in other Python files. Python has many <font color="orange">**built-in modules**</font> ready to use.
<br><br><br><br>
___

#### 2.3.1 Introduction to Modules

Some useful Python built-in modules:
- <font color="orange">**math**</font>: for mathematical operations
- <font color="orange">**random**</font>: for random numbers
- <font color="orange">**datetime**</font>: for dates and times

```python
# Example 2.2.1: Using the math Module
import math
```

```python
# math.sqrt function for square root
result = math.sqrt(16)

print("Square root of 16:", result)
```

```python
# Constant pi
circumference = 2 * math.pi * 5
print("Circumference of circle with radius 5:", circumference)
```

<br><br>
#### 2.3.2 Random Module

The <font color="orange">**random**</font> module is used to generate random numbers or choices.

```python
# Example 2.2.2: Using the random Module
import random
```

```python
# Random number between 1 and 10
random_number = random.randint(1, 10)
print("Random number:", random_number)
```

```python
# Choose random element from choices
choices = ["Rock", "Scissors", "Paper"]
result = random.choice(choices)
print("Choice result:", result)
```

<br><br>
#### 2.3.3 Datetime Module

The <font color="orange">**datetime**</font> module is used to work with dates and times in Python.

```python
# Example 2.3.3: Using the datetime Module
from datetime import datetime
```

```python
# Get current date and time
current_time = datetime.now()
print("Current date and time:", current_time)
```

```python
# Get only the year, month, and day
today = datetime.now()
print("Today's date:", today.year, "-", today.month, "-", today.day)
```

```python
# Create a specific date
birthday = datetime(2000, 5, 15)
print("Birthday:", birthday)
```
